```python
# Ejercicio 3

x = float(input('Ingrese primer número'))
y = float(input('Ingrese segúndo número'))

if x + y < 200:
   print('Menor a 200')

elif 200 < x + y < 250:
   print('Mayor a 200')

else:
   print('Mayor a 250')

```

```python
# Ejercicio 4

edad = float(input('¿Cuántos años tienes? '))
gusta_musica_urbana = input('¿Te gusta la música urbana? (SI/NO): ').strip().upper()

if edad >= 18:
    if gusta_musica_urbana == "SI":
        print("Eres mayor de edad y te gusta la música urbana.")
    else:
        print("Eres mayor de edad y no te gusta la música urbana.")
else:
    if gusta_musica_urbana == "SI":
        print("Eres menor de edad y te gusta la música urbana.")
    else:
        print("Eres menor de edad y no te gusta la música urbana.")
```
